```python
import pandas as pd
import matplotlib.pyplot as plt # type: ignore

df = pd.read_csv('conversion_log.csv')
# df.head(5)
conversion_counts = df['Type'].value_counts()
conversion_counts.plot(kind='bar')
plt.title('Number of Conversions per Type')
plt.xlabel('File Type')
plt.ylabel('Number of Conversions')
plt.show()

plt.show()
# plt.savefig('conversions_per_type.png')
```

```python
conversion_counts = df['Type'].value_counts()

# Create a bar chart
conversion_counts.plot(kind='pie')
plt.title('Number of Conversions per Type')
plt.xlabel('File Type')
plt.ylabel('Number of Conversions')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Number of Pages'])
plt.title('Number of Pages for Each Conversion')
plt.xlabel('Conversion Index')
plt.ylabel('Number of Pages')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Conversion Time (seconds)'])
plt.title('Required time for Each Conversion')
plt.xlabel('Conversion Index')
plt.ylabel('Time(Seconds)')
plt.show()
```

```python
plt.plot( df.index, (df['Conversion Time (seconds)']/df['Number of Pages']))
plt.title('Each page conversation time')
```
